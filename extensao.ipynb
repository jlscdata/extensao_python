```python
# Remoção de linhas com valores nulos
df.dropna(inplace=True)
```

```python
# Visualização das primeiras 20 colunas 
df.iloc[:, :20]
```

```python
# Comissão de cada vendedor
comissao_vendedor = df.groupby('NOME VENDEDOR')['Valor Comissão'].sum().reset_index()
comissao_vendedor = comissao_vendedor.sort_values(by='Valor Comissão', ascending=False)
print(comissao_vendedor)
```

```python
# Representação gráfica da comissao por vendedor
plt.figure(figsize=(10, 6))
barplot = sns.barplot(x='Valor Comissão', y='NOME VENDEDOR', data=comissao_vendedor, palette='viridis')
plt.title('Comissão Total por Vendedor')
plt.xlabel('Comissão Total')
plt.ylabel('Vendedor')
plt.show()
```

```python
# Porcentagem de venda

contagem_banco = df['BANCO'].value_counts()
porcentagem_banco = (contagem_banco / contagem_banco.sum()) *100
resultado = porcentagem_banco.reset_index()
resultado.columns = ['Banco' , 'Porcentagem']
print(resultado)
```

```python
# Visualização grafica de vendas por banco

plt.figure(figsize=(10, 6))
plt.bar(resultado['Banco'], resultado['Porcentagem'], color='red')
cores = plt.cm.viridis(np.linspace(0, 1, len(resultado)))
plt.title('Porcentagem de Vendas por Banco')
plt.xlabel('Banco')
plt.ylabel('Porcentagem (%)')
for index, value in enumerate(resultado['Porcentagem']):
    plt.text(index, value + 1, f'{value:.2f}%', ha='center')

plt.xticks(rotation=45)  
plt.ylim(0, resultado['Porcentagem'].max() + 10)  
plt.show()
```

```python
# Análise de gastos por cliente
gasto_cliente = df.groupby('NOME CLIENTE') ['VALOR VENDA'].sum().reset_index()
total_gasto = gasto_cliente['VALOR VENDA'].sum()
gasto_cliente['Porcentagem'] = (gasto_cliente['VALOR VENDA'] / total_gasto)*100
gasto_cliente['Porcentagem'] = gasto_cliente['Porcentagem'].round(2)
gasto_cliente = gasto_cliente.sort_values(by='VALOR VENDA', ascending=False)
print(gasto_cliente)
```
